```python
import torch
from diffusers import DiffusionPipeline
```

```python
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")
```

```python
prompt = "A breathtaking ultra-detailed 4K image of deep space — a wide-angle view of a vibrant nebula glowing in shades of purple, blue, and pink, with tendrils of cosmic dust swirling in intricate patterns. In the foreground, a large ringed planet with a glassy surface reflects distant stars. To the left, a futuristic spacecraft slowly glides past, leaving behind a soft plasma trail. Tiny asteroids and glowing stars of varying sizes scatter across the dark void. In the far background, a distant galaxy spirals, emitting a faint, ethereal glow. Rendered in cinematic lighting, ultra-realistic textures, ray-traced reflections, and fine cosmic particle effects. High contrast, deep blacks, vivid highlights — hyper-realistic space scene, 4K resolution."
images = pipe(prompt=prompt).images[0]
```
